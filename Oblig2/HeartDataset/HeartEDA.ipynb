# Feature Explanation

- Age
- Sex
- cp: Chest Pain Type
- trtbps: Resting Blood Pressure

- fbs: Fasting blood sugar
- restecg: Resting Electrocardiographic Results
- thalachh: Max Heart rate Achieved

- oldpeak: Previous peak
- slp: Slope
- caa: Number of major vessels
- thall: Thallium Stress Test

# EDA

### Distribution of categorical features

```python
fig = plt.figure(figsize=(32,16))
gs = fig.add_gridspec(3,3)
plot1 = fig.add_subplot(gs[0,0])
plot2 = fig.add_subplot(gs[0,1])
plot3 = fig.add_subplot(gs[0,2])
plot4 = fig.add_subplot(gs[1,0])
plot5 = fig.add_subplot(gs[1,1])
plot6 = fig.add_subplot(gs[1,2])
plot7 = fig.add_subplot(gs[2,0])
plot8 = fig.add_subplot(gs[2,1])
plot9 = fig.add_subplot(gs[2,2])

sns.countplot(ax=plot1,x=df['sex'])
plot1.set_title('Sex')

sns.countplot(ax=plot2,x=df['cp'])
plot2.set_title('Chest Pain Type')

sns.countplot(ax=plot3,x=df['fbs'])
plot3.set_title('Fasting Blood Sugar')

sns.countplot(ax=plot4,x=df['restecg'])
plot4.set_title('Resting Electrocardiographic Results')

sns.countplot(ax=plot5,x=df['exng'])
plot5.set_title('Exercise Induced Angina')

sns.countplot(ax=plot6,x=df['slp'])
plot6.set_title('Slope')

sns.countplot(ax=plot7,x=df['caa'])
plot7.set_title('Number of Major Vessels')

sns.countplot(ax=plot8,x=df['thall'])
plot8.set_title('Thalium Stress Test')

sns.countplot(ax=plot9,x=df['output'])
plot9.set_title('Output')
```

Here we can see the distribution of all the categorical features.

From these countplots we see that:
- There are more than twice has many with gender = 1 than gender = 0
- The ground truth is pretty balanced
- Most variables are not balanced
    - Could lead to problems when training

### Distribution of continuous features

```python
fig = plt.figure(figsize=(32,16))
gs = fig.add_gridspec(2,3)
plot1 = fig.add_subplot(gs[0,0])
plot2 = fig.add_subplot(gs[0,1])
plot3 = fig.add_subplot(gs[0,2])
plot4 = fig.add_subplot(gs[1,0])
plot5 = fig.add_subplot(gs[1,1])

sns.boxplot(ax=plot1,x=df['age'])
plot1.set_title('Age')

sns.boxplot(ax=plot2,x=df['trtbps'])
plot2.set_title('trtbps')

sns.boxplot(ax=plot3,x=df['chol'])
plot3.set_title('Cholesterol')

sns.boxplot(ax=plot4,x=df['thalachh'])
plot4.set_title('thalachh')

sns.boxplot(ax=plot5,x=df['oldpeak'])
plot5.set_title('oldpeak')
```

From these boxplots we can see that:
- All features contain outliers

```python
fig = plt.figure(figsize=(32,16))
gs = fig.add_gridspec(3,3)
plot1 = fig.add_subplot(gs[0,0])
plot2 = fig.add_subplot(gs[0,1])
plot3 = fig.add_subplot(gs[0,2])
plot4 = fig.add_subplot(gs[1,0])
plot5 = fig.add_subplot(gs[1,1])
plot6 = fig.add_subplot(gs[1,2])
plot7 = fig.add_subplot(gs[2,0])
plot8 = fig.add_subplot(gs[2,1])

sns.countplot(ax=plot1,x=df['sex'], hue=df['output'])
plot1.set_title('Sex')

sns.countplot(ax=plot2,x=df['cp'], hue=df['output'])
plot2.set_title('Chest Pain Type')

sns.countplot(ax=plot3,x=df['fbs'], hue=df['output'])
plot3.set_title('Fasting Blood Sugar')

sns.countplot(ax=plot4,x=df['restecg'], hue=df['output'])
plot4.set_title('Resting Electrocardiographic Results')

sns.countplot(ax=plot5,x=df['exng'], hue=df['output'])
plot5.set_title('Exercise Induced Angina')

sns.countplot(ax=plot6,x=df['slp'], hue=df['output'])
plot6.set_title('Slope')

sns.countplot(ax=plot7,x=df['caa'], hue=df['output'])
plot7.set_title('Number of Major Vessels')

sns.countplot(ax=plot8,x=df['thall'], hue=df['output'])
plot8.set_title('Thalium Stress Test')
```

From this countplot with regards to the ground truth, we can see that:
- Gender = 0, had more cases of heart failure than without
- Chest Pain Type = 2, had an unproportionate number of heart failures
- Fasting Blood Sugar has no correlation with heart failure
- Restecg = 1, has a slightly unproportionate amount of heart failures
- Most heart failure were not due to exercise, meaning heart failures mostly comes from genetics and life style
- Slope = 2, has an unproportionate amount of heart failures
- Number of major vessels = 0, has the greatest chance of heart failure in this feature
- Thallium Stress test = 2, has the greatest chance of heart failure in this feature

```python
corr = df.corr()
plt.figure(figsize=(32, 12))
heatmap = sns.heatmap(corr, annot=True)
```

Probably the easiest way to find correlations between variables are with a correlation heatmap.
Here we can see a great correlation between the ground truth and:
    - Age
    - Sex
    - cp
    - trtbps
    - restecg
    - thalachh
    - exng
    - oldpeak
    - slp
    - caa
    - thall

However, chol and fbs has little to no correlation with the ground truth. If this dataset had been bigger it could have been an idea to remove these columns to reduce dimensionality.
