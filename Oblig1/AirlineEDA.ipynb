## Null-values

```python
df.isnull().sum()
```

```python
def handle_null_median(df):
    # Need to set inplace=True, so it doesn't create a copy of the dataframe. Tried without and this led to null-values not being removed
    df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median(), inplace=True)

    return  df

# Sets null values equal to the mean of that column
def handle_null_mean(df):
    df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean(), inplace=True)

    return  df

# Function to drop all rows with at least 1 null value. (Removes way to many rows, so isn't a good option)
def null_values_drop(df):
    df_copy = df.copy()
    df_drop = df_copy.dropna(axis=0, how='any')

    return df_drop

df = handle_null_mean(df)
df.head()
```

Here I've created three functions to handle missing value for the given column. All three strategies would work fine on this dataset since it so big and so few missing value.
I chose to handle the missing value with the mean value for that row, but I could might as well have used one of the other

## Feature Engineering

```python
df['Late Departure'] = 0
df.loc[df['Departure Delay in Minutes'] != 0, 'Late Departure' ] = 1
df = df.drop('Departure Delay in Minutes', axis=1)

df['Late Arrival'] = 0
df.loc[df['Arrival Delay in Minutes'] != 0, 'Late Arrival' ] = 1
df = df.drop('Arrival Delay in Minutes', axis=1)

df.head()
```

Created two categorical columns: Late Departure and Late Arrival
I've created these to see if it matters how late the plain arrived or departed.

```python
pd.cut(df['Flight Distance'], 10)
```

```python
# Got the idea of splitting age into bins from: https://www.kaggle.com/code/sinakhorami/titanic-best-working-classifier, but seems like a common practice called Binning, where you put a range of values into a bin
def bins(df): # This is kind of feature engineering as well
    df['FlightDistanceGroup'] = 0
    df.loc[df['Flight Distance'] < 526.2, 'FlightDistanceGroup' ] = 0
    df.loc[(df['Flight Distance'] >= 526.2) & (df['Flight Distance'] < 1021.4), 'FlightDistanceGroup'] = 1
    df.loc[(df['Flight Distance'] >= 1021.4) & (df['Flight Distance'] < 1516.6), 'FlightDistanceGroup'] = 2
    df.loc[(df['Flight Distance'] >= 1516.6) & (df['Flight Distance'] < 2011.8), 'FlightDistanceGroup'] = 3
    df.loc[(df['Flight Distance'] >= 2011.8) & (df['Flight Distance'] < 2507.0), 'FlightDistanceGroup'] = 4
    df.loc[(df['Flight Distance'] >= 2507) & (df['Flight Distance'] < 3002.2), 'FlightDistanceGroup'] = 5
    df.loc[(df['Flight Distance'] >= 3002.2) & (df['Flight Distance'] < 3497.4), 'FlightDistanceGroup'] = 6
    df.loc[(df['Flight Distance'] >= 3497.4) & (df['Flight Distance'] < 3992.6), 'FlightDistanceGroup'] = 7
    df.loc[(df['Flight Distance'] >= 3992.6) & (df['Flight Distance'] < 4487.8), 'FlightDistanceGroup'] = 8
    df.loc[df['Flight Distance'] >=  4487.8, 'FlightDistanceGroup'] = 9

    df = df.drop(columns=['Flight Distance'])

    return df

df = bins(df)
df.head()
```

I've done this binning since this makes continious values to categorical ones. This makes it possible to look at the distribution of the column much easier. This also makes it possible to do feature correlation analysis.

## Multivariate Analysis

```python
corr = df.corr()
plt.figure(figsize=(32, 12))
heatmap = sns.heatmap(corr, annot=True)
```

The dataset contains way to many columns to look at correlations between each and every column, so I've chosen a few to take a look at.

From this correlation plot we can see:
- Gender has no clear correlation with any other feature
- Age has a positive correlation with Online boarding. Meaning has a person gets older, the higher it scores online boarding
- Inflight wifi service and ease of online booking has a great positive correlation. Meaning people who scored ease of booking online high, also scored inflight wifi services high
- The higher ticket class(higher is worse) the worse one generally would score other features.
- If a person had high seat comfort they would one average score more features high. The same goes for inflight entertainment.
- Most features are highly correlated with the ground truth.
- My two feature engineered features has close to zero correlation with satisfaction. This indicates that the amount of minutes a plane arrives or departs late makes a big difference for people.

For a lot of features, if a person scored one feature really high, it seems like the person was more likely to score other features high, especially features which are semi in the same category.

From this correlation heatmap I could safely remove these columns due to a lack of correlation iwth other features and the ground truth:
- Gender
- Late arrival
- Late departure

No idendpendent variable has such a high correlation with another independent variable that I could remove one of them without loosing information.
