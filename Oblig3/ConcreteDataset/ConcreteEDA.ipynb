# EDA

### Distribution of features

```python
fig = plt.figure(figsize=(32,16))
gs = fig.add_gridspec(3,3)
plot1 = fig.add_subplot(gs[0,0])
plot2 = fig.add_subplot(gs[0,1])
plot3 = fig.add_subplot(gs[0,2])
plot4 = fig.add_subplot(gs[1,0])
plot5 = fig.add_subplot(gs[1,1])
plot6 = fig.add_subplot(gs[1,2])
plot7 = fig.add_subplot(gs[2,0])
plot8 = fig.add_subplot(gs[2,1])
plot9 = fig.add_subplot(gs[2,2])

sns.boxplot(ax=plot1,x=df['Cement (component 1)(kg in a m^3 mixture)'])
plot1.set_title('Cement')

sns.boxplot(ax=plot2,x=df['Blast Furnace Slag (component 2)(kg in a m^3 mixture)'])
plot2.set_title('Blast Furnace Slag')

sns.boxplot(ax=plot3,x=df['Fly Ash (component 3)(kg in a m^3 mixture)'])
plot3.set_title('Fly Ash')

sns.boxplot(ax=plot4,x=df['Water  (component 4)(kg in a m^3 mixture)'])
plot4.set_title('Water')

sns.boxplot(ax=plot5,x=df['Superplasticizer (component 5)(kg in a m^3 mixture)'])
plot5.set_title('Superplasticizer')

sns.boxplot(ax=plot6,x=df['Coarse Aggregate  (component 6)(kg in a m^3 mixture)'])
plot6.set_title('Coarse Aggregate')

sns.boxplot(ax=plot7,x=df['Fine Aggregate (component 7)(kg in a m^3 mixture)'])
plot7.set_title('Fine Aggregate')

sns.boxplot(ax=plot8,x=df['Age (day)'])
plot8.set_title('Age')

sns.boxplot(ax=plot9,x=df['Concrete compressive strength(MPa, megapascals) '])
plot9.set_title('Concrete compressive strength')
```

Most features are pretty evenly distributed, but some features has some major outliers. This could be a problem for some algorithms.
