```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import normalize
import tensorflow as tf
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Dense, Input, Dropout
```

```python
df = pd.read_csv(r'C:\Users\marku\Desktop\ML\MLGit\datasets\Star.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
features = df.drop('Star type',axis=1)
targets = df['Star type']

targets_onehot = pd.get_dummies(targets)
features_onehot = pd.get_dummies(features)
targets_onehot = targets_onehot.astype('float32')
features_onehot = features_onehot.astype('float32')

features_onehot.head()
```

```python
features_onehot = normalize(features_onehot)

train_features, test_features, train_targets, test_targets = train_test_split(features_onehot, targets_onehot, test_size=0.25, random_state=42)
```

```python
train_features_tensor = tf.convert_to_tensor(train_features)
test_features_tensor = tf.convert_to_tensor(test_features)
train_targets_tensor = tf.convert_to_tensor(train_targets)
test_targets_tensor = tf.convert_to_tensor(test_targets)
```

# NN Architecture 1

```python
model1 = Sequential()

model1.add(Input(30))

model1.add(Dense(60, activation='relu'))
model1.add(Dense(60, activation='relu'))
model1.add(Dense(30, activation='relu'))

model1.add(Dense(6, activation='sigmoid'))
```

```python
opt1 = Adam()
model1.compile(optimizer=opt1, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model1.fit(train_features_tensor, train_targets_tensor, epochs=500, batch_size=10)
```

```python
model1.evaluate(train_features_tensor, train_targets_tensor)
```

```python
model1.evaluate(test_features_tensor, test_targets_tensor)
```

At first I had problems with the test accuracy outperforming the training accuracy. After changing the percent used for training/test this was mostly fixed. Then the NN was overfitting. Will try to fix this in the next architecture

# NN Architecture 2

```python
model2 = Sequential()

model2.add(Input(30))

model2.add(Dense(60, activation='relu'))
model2.add(Dropout(0.1))

model2.add(Dense(60, activation='relu'))
model2.add(Dropout(0.4))

model2.add(Dense(30, activation='relu'))
model2.add(Dropout(0.1))

model2.add(Dense(30, activation='relu'))
model2.add(Dense(6, activation='sigmoid'))
```

```python
opt1 = Adam()
model2.compile(optimizer=opt1, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model2.fit(train_features_tensor, train_targets_tensor, epochs=500, batch_size=10)
```

```python
model2.evaluate(train_features_tensor, train_targets_tensor)
```

```python
model2.evaluate(test_features_tensor, test_targets_tensor)
```
